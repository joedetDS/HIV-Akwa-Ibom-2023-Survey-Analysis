##### In the 'positve status year' column, we try to remove the nan and 'None' and change it to ), for better understanding. We also convert the entire column into year this helps the column to have a generic format.

```python
hiv_df['positive_status_year'].replace([np.nan,'None'],'0',inplace=True)
```

```python
#List of some keys that indidcates this year

values_to_change=['Not long ','Of recent ', 'Two weeks', '5 months','6 months']  
```

```python
#reolace the keys with this year

hiv_df['positive_status_year'].replace(values_to_change,'2023',inplace=True)
```

```python
#View the keys

hiv_df['positive_status_year'].unique()
```

##### We have to remove the 'years' in the records and leave only the year and also create a function to subtract the current year fromt he numerical value to give us the year the individual found out he/she tested positive.

```python
##Function to split

def split(x):
    k=str(x).split()
    return k[0].strip('years')
```

```python
#Apply the function to the column

hiv_df['positive_status_year']=hiv_df['positive_status_year'].apply(split)
```

```python
#function to display the actual year

def convert_to_year(x):
    if x==str(0):
        return 0
    if x=='2023':
        return x
    if x!=0:
        return 2023-int(x)
```

```python
#Apply the function

hiv_df['positive_status_year']=hiv_df['positive_status_year'].apply(convert_to_year)
```

```python
#Perform a value count

hiv_df['positive_status_year'].value_counts()
```

##### From this change, we can go on to create two new column to indicate those positive and negative records and the second one to indicate year since he/she was tested positive. We use 'Yes' and 'No' as records to indicate the differences in the first column.

```python
#Function for year

def HIV_positive(year):
    if year == 0:
        return "No"
    else:
        return "Yes"
```

```python
#Create the year_since function

def year_since(year):
    if year == 0:
        return 0
    else:
        return 2023-int(year)
```

```python
#Apply the function

hiv_df["hiv_positive"]=hiv_df['positive_status_year'].apply(HIV_positive)
```

```python
##Apply the function

hiv_df['year_since_positive']=hiv_df['positive_status_year'].apply(year_since)
```

```python
#Fill the remaining null values with not stated

hiv_df.fillna('Not stated',inplace=True)
```

```python
#Replace Option1 with Others

hiv_df.replace('Option 1','Others',inplace=True)
```

##### In the Support desc column, some user input are synonymous, we can go ahead to replace similar words with a generic word. This will help if we are to carry further analysis like visualization. Words with check up could just be written as check up and words containig drugs could be replaced to more drugs

```python
#Function to clean the column
def clean_support(word):
    if ("drugs" in word):
            if "check ups" in word:
                return "Check ups and more drugs"
            else:
                return "More drugs"
    else:
        return word
```

```python
#Apply the column

hiv_df['support_desc']=hiv_df['support_desc'].apply(clean_support)
```

```python
#Check if change has been effected

hiv_df['support_desc'].value_counts()
```

```python
#View the value counts of the column

hiv_df['additional_comments'].value_counts()
```

##### The additional comments column look somewhat similar, we pick common records and replace them with a generic word and also correct the spelling of some words.

```python
#Words synonymous to awareness 

awareness=['reaching','awareness','enlightness']
```

```python
#Function to effect changes

def  clean_add_comment(word):
    if ("descriminated" in word):
        return "Avoid Discrimination"
    elif (("awareness" in word) or ("enlightened" in word) or ("reaching" in word)):
        return "Create Awareness"
    elif "rate" in word:
        return "help from government"
    elif "drugs" in word:
        return "more drugs"
    else:
          return word
```

```python
#Apply function on the column

hiv_df['additional_comments']=hiv_df['additional_comments'].apply(clean_add_comment)
```
